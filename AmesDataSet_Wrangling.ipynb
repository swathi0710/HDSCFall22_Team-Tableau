```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
data = pd.read_csv("/content/train.csv")
data.info()
```

```python

data.select_dtypes("object").nunique().sort_values(ascending=False)
```

```python
data.drop(columns=["Exterior2nd","Exterior1st","MSZoning","Street","Alley","LotShape","LandContour","LotConfig","LandSlope"],inplace=True)
```

```python

data.select_dtypes("object").nunique().sort_values(ascending=False)
```

```python
na=data.isna().sum().sort_values(ascending=False)
na[na>0]
 
```

```python
cat=data.select_dtypes("object")
[data[x].replace(np.nan, 'NA', inplace=True) for x in cat.columns]
```

```python
num_data=data.select_dtypes("int64")
[data[col].replace(np.nan,'0.0', inplace=True) for col in num_data]
```

```python
#data.drop("Id", axis = 1, inplace = True)
data.fillna(0) #for LotFrontage which is a float type
```

```python
data.isnull().sum() 
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE
from sklearn.tree import DecisionTreeRegressor

X=data.drop(["SalePrice"],axis=1).fillna(0).select_dtypes(exclude='object')
Y=data["SalePrice"]
mod=DecisionTreeRegressor()
rfe = RFE(estimator=mod, n_features_to_select=20)
features=rfe.fit(X,Y)
Res = list(X.select_dtypes(exclude='object').columns[features.support_])
Res
Res.append("SalePrice")
```
